```python
import requests # basit http istekleri oluşturmak için.
```

```python
import os # işletim sistemi ile dosya işlemleri yapmak için; yol öğrenme, dosya oluşturma vs.
```

```python
from tqdm import tqdm # progress bar için.
```

```python
from bs4 import BeautifulSoup as bs # html ayıklarken basitleştirici bir kütüphane.
```

```python
from urllib.parse import urljoin, urlparse # tam yol belirleme, url ayrıştırma.
```

```python
def is_valid(url):
    """
    Url doğrulaması.
    """
    parsed = urlparse(url)
    return bool(parsed.netloc) and bool(parsed.scheme)
```

```python
def get_all_images(url):
    """
    Verilen urldeki tüm resimlerin urlsini döndürür.
    """
    soup = bs(requests.get(url).content, "html.parser") # çorba yap
    urls = []
    for img in tqdm(soup.find_all("img"), "Slider görseller bulunuyor"):
        img_url = img.attrs.get("src")
        if img_url.find("slide") == -1:
            # slide kelimesi geçenleri ayıkla.
            continue
        if not img_url:
            # src kısmı yoksa geç.
            continue
        img_url = urljoin(url, img_url) # imajın tam yolunu al.
        try:
            # url temizliği
            pos = img_url.index("?")
            img_url = img_url[:pos]
        except ValueError:
            pass
        if is_valid(img_url):
            # url geçerli ise urlyi listeye al.
            urls.append(img_url)
    return urls
```

```python
def download(url, pathname):
    """
    Verilen urlyi indirir, verilen uzantıya koyar.
    """
    if not os.path.isdir(pathname):
        # eğer directory yoksa oluşturur.
        os.makedirs(pathname)
    
    # indirme sırasında işlem yapmak için stream true veriyoruz.
    response = requests.get(url, stream=True) 

    # indirilecek dosyanın boyutunu alalım.
    file_size = int(response.headers.get("Content-Length", 0))

    # ismini alalım.
    filename = os.path.join(pathname, url.split("/")[-1])

    # tqdm kütüphanesini kullanarak progress bar oluşturalım.
    progress = tqdm(response.iter_content(1024), f"İndiriliyor {filename}", total=file_size, unit="B", unit_scale=True,
                    unit_divisor=1024)
    
    with open(filename, "wb") as f:
        # write and binary modda açtık, okundukça yazacaz.
        for data in progress.iterable:
            f.write(data)  # okuduğunu yaz
            progress.update(len(data)) # ilerleme çubuğunu yenile.
```

```python
def main(url, path):
    imgs = get_all_images(url) # tüm urlleri al
    for img in imgs:
        download(img, path) # teker teker indir
```
